**Filter All The Things**

When you start to explore your data you are going to want to begin breaking it down into smaller chunks so you can wrap your head around trends and insights that are starting to form in your mind.

Being able to filter data will help you answer questions like these:

what are all the games older than 2002?
What games by Nintendo were published in 2008 and exceeded 3.5 mil in sales?



```
# people whose "age" is greater than 30

df[df["age"] > 30]
```

```
# people who have more pets than children

df[df["num_pets"] > df["num_children"]]
```

```
# people older than 40 who own pets

df[(df["age"] > 40) & (df["num_pets"] > 0)]
```

```
# people older than 40 or people who own pets
df[(df["age"] > 40) | (df["num_pets"] > 0)]
```
