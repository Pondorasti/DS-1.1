```python
import pandas as pd

data_url = "datasets/starbucks_drinkMenu_expanded.csv"
#source: https://www.kaggle.com/starbucks/starbucks-menu?select=starbucks_drinkMenu_expanded.csv

starbucks_dataframe = pd.read_csv(data_url)
starbucks_dataframe
```

Bar charts, otherwise known as bar graphs, bar plots, or histograms (you may have heard of those!) are very simple visualizations that allow you to plot categorical or quantitative data in a visually separable manner.

Let's start by playing with the simpler of the two to visualize with bar charts: categorical data.

Categorical Data: variables that represent data that can be divided into groups
Continuous Data: quantity 

```python
import seaborn as sns
from matplotlib import rcParams
#bar chart with categorical data

rcParams['figure.figsize'] = 20,8.27

#Beverage category counts for drinks with more than 100 calories?
drinks_greater_100_df = starbucks_dataframe[starbucks_dataframe["Calories"] > 100]

sns.countplot(x="Beverage_category", palette="spring", data=drinks_greater_100_df)
```

```python
#what about continuous data?
starbucks_dataframe["Calories"].hist(bins=16)
```

```python
#starbucks_dataframe[" Sugars (g)"]

sns.barplot(x="Beverage_category", y=" Sugars (g)", data=starbucks_dataframe)

starbucks_dataframe = starbucks_dataframe.dropna()
# starbucks_dataframe = starbucks_dataframe[(starbucks_dataframe['Caffeine (mg)'] != 'Varies') & (starbucks_dataframe['Caffeine (mg)'] != 'varies')]
starbucks_dataframe = starbucks_dataframe[starbucks_dataframe['Caffeine (mg)'].astype(str).str.isnumeric()]

def toInt(num):
  return int(num)

# starbucks_dataframe['Caffeine (mg)'] = starbucks_dataframe['Caffeine (mg)'].apply(toInt)
drinks_with_Caffeine = starbucks_dataframe[starbucks_dataframe['Caffeine (mg)'] > 50]

sns.barplot(x='Beverage_category',y='Caffeine (mg)', data=drinks_with_Caffeine)
```
