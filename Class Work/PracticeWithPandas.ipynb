A dataframe is one of the most important data structures included in the pandas library. You can think of dataframes as fancy excel tables. They store data in row and column format and let you easily manipulate large amounts of data. 

pandas comes with a built in function called read_csv() that let's use easily convert a CSV file to a dataframe

For info about read_csv() check out [this article](https://www.datacamp.com/community/tutorials/pandas-read-csv)

loc and iloc are useful methods for selecting columns of a data frame based on label in the case of loc and integer position in the case of iloc (note that both use bracket notation). To make the difference clear we will use set_index to make the "id" column our index labels and then we can use loc to select specific rows or slices of rows based on label or iloc to select based on integer position. Read more about loc and iloc [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html)

````
df = df.set_index("id")

#TODO: select the row labeled 282614216
df.loc[282614216]
#TODO: select the 20th row using iloc
df.iloc[19]
````

You may have noticed that the resulting object type of loc and iloc is not a dataframe but a series, a series is a 1-d labeled array, dataframes are acutally composed of series, with each row and column being a series, for more info on dataframes and series read [this article](https://towardsdatascience.com/pandas-series-dataframe-explained-a178f9748d46)

Sometimes we may want to select data based on some sort of boolean condition, for example we want all the apps from the "Shopping" genere. To do this we can use basic comparison operators < > <= >= == and boolean operators & (and) | or to select the desired data. 

[This article](https://towardsdatascience.com/filtering-data-frames-in-pandas-b570b1f834b9) is a great summary of all the selection techniques (boolean selection towards the end)

````
df[df["prime_genre"] == "Shopping"]
#TODO: select all the apps that are more than 3 dollars
df[df["price"] > 3]
#TODO: select all the apps that have less than 30 ratings for the current version or are rated less than 3.0
df[(df["rating_count_tot"] < 30) | (df["user_rating"] < 3.0)]
````

apply is used to apply a function efficiently to rows in a dataframe, it's useful to generating new columns of data or for formatting data nicely, for example you could apply to make a new column that expresses the bytes column in terms of megabytes to make it more readable, check out [this article](https://towardsdatascience.com/introduction-to-pandas-apply-applymap-and-map-5d3e044e93ff) for some nice visuals and explanation of apply()

````
import numpy as np #we will learn more about numpy soon!
def byte_resizer(data):
    return np.around(data / 1000000, decimals=2)

df["size_Mb"] = df["size_bytes"].apply(byte_resizer)

#You can also use some built in functions to peform certain functions on columns
#let's say we want to get the average number of bytes:

df["size_bytes"].mean()
````

Groupby is used to easily split data into groups. For example let's say we want to look at the average byte size for each genre, we could do that using the code below: 

check out [this article](https://naomi-fridman.medium.com/pandas-groupby-explained-with-titanic-6a1c47eb8182) for more detailed information on how groupby works!

````
df.groupby("prime_genre")["size_bytes"].hist()
````

We can combine a number of the commands we have learned so far to find the answers to even more interesting questions! For example here's one way we could combine several commands to see the average current version rating for apps costing more than $20 in each genre!

````
df[(df["price"] > 20.0) & (df["currency"] == "USD")].groupby("prime_genre")["user_rating_ver"].mean()
````
